# Religion, dealing with missing values

The first thing is to get the  details in the religion colunm to determine what should be done.
Religion for minors should be replaced with their parent's because in most cases a child grow to practice the relation of their parent.Also, a child of even 12 years has already build his belive arround the parent/guidance religion there are not yet of age to chose their religion.In most cases that a person convert to another religion is always in thier 30s and 40s were they would have learn something differently.So I imputed the minors religion to that of the previous respondent after them(using time series) which is likely their parent/guidance since they cannot be Head Of House.

Another issues with religion is error made by some respondents.There is an instance were someone inputted Christ as their religion. This should be replaced with Christian as there is no religion like 'Christ'. The followers of Christ are called 'Christians'.

Also, some branches of Christianity such as Baptist,Orthodoxy,Methodist,Catholic should be replace with 'Christian'. But for the sake of decision making on the piece of land(if to build a church and what denomination), I will leave Methodist and Catholic since the have high number of members.

A respondence gave his religion as 'Undecided'. This will be replaced with 'None' as it is clear that he does not believe in God but does not just want to admit it. If atheism was a religion, it would have been replaced with it.Private too will be replaced with 'None'

The blanks in this colunm will be replaced with 'NaN'

Other imputations will be done in the course of the analysis


```python
unique_Religion= census['Religion'].unique()
unique_Religion
```

```python
# imputation of missing data
census['Religion'].replace(['Undecided'],['None'], inplace = True)
census['Religion'].replace(['Private'],['None'], inplace = True)
census['Religion'].replace(['Cath0lic'],['Christian'], inplace = True)
census['Religion'].replace(['Orthodoxy'],['Christian'], inplace = True)
census['Religion'].replace(['Christ'],['Christian'], inplace = True)
census['Religion'].replace(['Baptist'],['Christian'], inplace = True)
census['Religion'].replace([''],['None'], inplace = True)
census.at[10162,'Religion']='Christian'
census['Religion'].replace([" "], ["NaN"], inplace = True)
census['Religion'].fillna(method='ffill',inplace=True)
print(census['Religion'].value_counts())
```

# Handling Outliers

Since we need to make decision based on analysisng the data, outrageous values must be identified and dealt with as it can significantly affect conclusions drawn.Outliers could be as result of error during data gathering or outright lies. For the purpose of this assignment, outliers were identified by visualization using Boxplot and mathematically by using Z-score.

Z score can used to identify outlier by showing an uncommon value in a dataset through signposting how far a data is from the mean.This is done by subtracting the mean and dividing by the STD.Z-scores of +/-3 or more from zero is the typical cut-off value for identifying outliers(https://dataclimbers.com/2021/01/27/ways-to-detect-outliers-in-dataset-using-python-and-pandas-exploratory-data-analysis/),01-12-2021.

By using boxplot for the Marital Status, it identified some widows at very young age as outlier especially Men.While I think that some divorced young people provided their Marital Status as widowed, I am also aware that it is not out of place to be widowed even at Age 18. 
For the Age ,I set a threshold value of 3 which implies that any value presented by Z score below 3 and above 3 is identified as an outlier.People with the ages 101,102,103,104 and 105 where identified as outlier, but I had to ingnore these because that is their  actual ages after thorough review of their other data such occupation(all Retired) and Relationship to Head of House.

#  How to calculate birth rate ( crude birth rate)
Divide total no of births in 1 year by the absoulte population and then multipy the by 1000


From the anaysis of population in the city, the birth rate for the current year is 10.12 while the previous year is 14.64which indicate a decline of 4.52  .The Reason for this decline could be as result of economic development,improved education for women,anxiety over future,change in social attitude, cost of bring children up, parenting less attractive or access to contraception(https://www.economicshelp.org/blog/167046/economics/reasons-for-a-falling-birth-rate/),07-12-2021

```python
Population = len(census)
births_one_year = len(census[census['Age'] == 0])
Birthrate = (births_one_year/Population)*1000
Birthrate
```

```python


age0 = census['Age'] == 0
age4= census['Age'] == 4
age8 = census['Age'] == 8

cbirthrate = (len(census[age0])/Population)*1000 

cbirthrate4 = (len(census[age4])/Population)*1000

print(f"The Birthrate per 1000 for the year is {cbirthrate}")
print(f"The Birthrate for the previous year per 1000 is {cbirthrate4}")
print(f"The difference between the previous Birthrate and the current year per 1000 is {cbirthrate-cbirthrate4}")
```

#  How to calculate death rate
Divide total no of deaths in 1 year by the absoulte population and then multipy the by 1000


A review of the death rate shows the rate of 11.99 which is below the world average as at the time of the census(https://www.macrotrends.net/countries/WLD/world/death-rate),07-12-2021 . This decline could be as result of accessibility to high quality health care even by the middele class.However, there is need for the city to plan for the elderly as this will lead to higher life expectancy which implies more aged people and increase in the dependent population who may not be able to take care of themselves.Creation of carehome should be one of the priorities of the decision makers.

```python
aging_population = census[census['Age'].between(65,113)].groupby('Age').size().sort_index(ascending=False)
ageing = aging_population.shift(-1)
net = aging_population - ageing
net = net.reset_index(name='Count')
deaths = net[net['Count'] < 0]
death_rate = abs(((deaths['Count']).sum())/Population)*1000
death_rate
```
